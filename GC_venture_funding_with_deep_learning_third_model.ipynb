```python
# Imports
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
# Upload credit_card_transactions.csv to Colab
from google.colab import files

csv_file = files.upload()
```

```python
# Read the applicants_data.csv file from the Resources folder into a Pandas DataFrame
applicant_data_df = pd.read_csv("applicants_data.csv")

# Review the DataFrame
applicant_data_df.head()
```

```python
# Review the data types associated with the columns
applicant_data_df.dtypes
```

```python
# Drop the 'EIN' and 'NAME' columns from the DataFrame DROP THE SEPCIAL SPECIAL_CONSIDERATIONS TOO!!!!!
applicant_data_df = applicant_data_df.drop(columns = ["EIN","NAME", "SPECIAL_CONSIDERATIONS"])
# Review the DataFrame
applicant_data_df
```

```python
# Create a list of categorical variables 
categorical_variables = list(applicant_data_df.dtypes[applicant_data_df.dtypes == "object"].index)

# Display the categorical variables list
categorical_variables
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse = False)
```

```python
# Encode the categorcal variables using OneHotEncoder
encoded_data = enc.fit_transform(applicant_data_df[categorical_variables])
```

```python
# Create a DataFrame with the encoded variables
encoded_df = pd.DataFrame(encoded_data,
    columns = enc.get_feature_names(categorical_variables)
)

# Review the DataFrame
encoded_df.head()
```

```python
# Add the numerical variables from the original DataFrame to the one-hot encoding DataFrame
numerical_variables_df = applicant_data_df.drop(columns = categorical_variables)

# Review the DataFrame
display(numerical_variables_df.head())

# Combine both data frames into one Data Frame
encoded_df = pd.concat(
    [
        numerical_variables_df,
        encoded_df
    ],
    axis=1
)

# Review the new data frame all together
display(encoded_df.head())
```

```python
# Define the target set y using the IS_SUCCESSFUL column
y = encoded_df["IS_SUCCESSFUL"]

# Display a sample of y
y[0:5]
```

```python
# Define features set X by selecting all columns but IS_SUCCESSFUL
X = encoded_df.drop(columns= ["IS_SUCCESSFUL"])

# Review the features DataFrame
X.head()
```

```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to the features training dataset
X_scaler = scaler.fit(X_train)

# Fit the scaler to the features training dataset
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the the number of inputs (features) to the model
number_input_features = len(X_train_scaled[0])

# Review the number of features
number_input_features
```

```python
# Define the number of neurons in the output layer
number_output_neurons_A3 = 1
```

```python
# Define the number of hidden nodes for the first hidden layer
hidden_nodes_layer1_A3 = (number_input_features + 1) // 2 

# Review the number of hidden nodes in the first layer
hidden_nodes_layer1_A3
```

```python
# Define the number of hidden nodes for the second hidden layer
hidden_nodes_layer2_A3 = (hidden_nodes_layer1_A3 + 1) // 2

# Review the number of hidden nodes in the second layer
hidden_nodes_layer2_A3
```

```python
# Define the number of hidden nodes for the third hidden layer
hidden_nodes_layer3_A3 = (hidden_nodes_layer2_A3 + 1) // 2

# Review the number of hidden nodes in the second layer
hidden_nodes_layer3_A3
```

```python
# Create the Sequential model instance
nn_A3= Sequential()
```

```python
# First hidden layer
nn_A3.add(Dense(units = hidden_nodes_layer1_A3, input_dim= number_input_features, activation = 'relu'))

# Second hidden layer
nn_A3.add(Dense(units = hidden_nodes_layer2_A3,activation = 'relu'))

# Thrid hidden layer
nn_A3.add(Dense(units = hidden_nodes_layer3_A3,activation = 'relu'))

# Output layer
nn_A3.add(Dense(units = number_output_neurons_A3, activation = 'sigmoid'))


# Check the structure of the model
nn_A3.summary()
```

```python
# Compile the model
nn_A3.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model
fit_model_A3 = nn_A3.fit(X_train_scaled, y_train, epochs=100)
```

```python
print("Alternative Model 3 Results")

# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = nn_A3.evaluate(X_test_scaled, y_test, verbose=2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
